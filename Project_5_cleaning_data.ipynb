Checking null value percentage

```python
null_perc = df.isnull().sum()/df.shape[0]*100
null_perc
```

```python
null_value_per_greater_than_50 = null_perc[null_perc > 50]
null_value_per_greater_than_50
```

Dropping columns in which null values percentage is greater than 50

```python
df.drop(columns = null_value_per_greater_than_50.index,inplace = True)
df.shape
```

```python
df.isnull().sum()
```

Seperating categorical column in one list

```python
categorical_col = df.select_dtypes(include = 'object')
categorical_col.head()
```

Seperating numerical col in one list

```python
numerical_col = df.select_dtypes(exclude= 'object')
numerical_col.head()
```

```python
categorical_col.isnull().sum()
```

Filling missing values of categorical column by mode

```python
for col in categorical_col:
  df[col] = df[col].fillna(df[col].mode()[0])
  print(df[col].mode()[0])
```

Filling missing values of numerical columns by mean

```python
for col in numerical_col:
  df[col] = df[col].fillna(df[col].mean())
  print(df[col].mean())
```

Checking null values after filling it

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

Checking outliers of columns

```python
for col in numerical_col:
  sns.boxplot(df[col])
  plt.show()
```

Handling outliers for all the columns except Datatime columns

```python
for col in numerical_col:
  if df[col].dtypes != 'datetime64[ns]':
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5*IQR
    upper_limit = Q3 + 1.5*IQR
    df[col] = np.where(df[col] > upper_limit,upper_limit,df[col])
    df[col] = np.where(df[col]< lower_limit , lower_limit,df[col])
    sns.boxplot(df[col])
    plt.show()
  else:
    print(f"skipping outlier for handling datetime column: {col}")
```
